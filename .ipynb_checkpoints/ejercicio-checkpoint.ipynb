Mi código crea un juego de bingo simple, primero creando una matriz (carton) con números al azar en un rango del 1 al 100, posteriormente elige un número al azar dentro del mismo rango. En el caso de que el número al azar escogido se encuentre dentro de la matriz creada anteriormente, se marca con una "x" verde y se visualiza con matplotlib.

# Paso 1:
El primer paso es importar las librerías necesarias:
Numpy
pandas
matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Paso 2
Se crea un cartón 5x5 con números aleatorios del 1 al 100 sin repetir.
Se genera un número al azar.
Se verifica si el número sorteado coincide con algún número del cartón.

```python
#Crear un cartón de bingo con números aleatorios del 1 al 100
carton = pd.DataFrame(np.random.choice(range(1, 101), (5, 5), replace=False))

#Sorteo de un número aleatorio del 1 al 100
numero = np.random.randint(1, 101)

print("Número sorteado:", numero)

#Marcar el cartón con una "X" si el número sorteado coincide con algún número del cartón
marcado = carton == numero 
```

# Paso 3
Convierte la matriz booleanas en valores numéricos (0 y 1) para visualización
Muestra el cartón
Se recorre cada celda con dos ciclos for 
Coloca una x donde el número coincide, mientras que muestra el número original cuando no coincide


```python
#Visualizar el cartón con los números marcados
plt.imshow(marcado.values.astype(int), cmap="Greens")

for i in range(5):
    for j in range(5):
        texto = "X" if marcado.iloc[i, j] else str(carton.iloc[i, j])
        plt.text(j, i, texto, ha='center', va='center')

plt.xticks([])
plt.yticks([])
plt.title("Bingo simple")
plt.show()
```

# Conclusiones 
Se hace uso de las 3 librerias mencionadas.
Con numpy se eligen los números aleatoriamente y crea la matriz.
Con Pandas se crea el dataframe a partir de la matriz creadacon numpy del cartón y se administra
Con matplotlib se grafica y se muestra visualmente el juego.
